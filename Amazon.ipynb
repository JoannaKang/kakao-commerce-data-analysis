```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```
amz_business_report_base_dir = "C:\\Users\\Joanna Kang\\Desktop\\TASK\\5. Data Anaylsis\\2020_July_trend\\Amazon\\businessreport"
amz_ad_report_base_dir = "C:\\Users\\Joanna Kang\\Desktop\\TASK\\5. Data Anaylsis\\2020_July_trend\\Amazon\\adreport"
amz_business_report_name = os.listdir("C:\\Users\\Joanna Kang\\Desktop\\TASK\\5. Data Anaylsis\\2020_July_trend\\Amazon\\businessreport")
amz_ad_report_name = os.listdir("C:\\Users\\Joanna Kang\\Desktop\\TASK\\5. Data Anaylsis\\2020_July_trend\\Amazon\\adreport")

amz_business_report_full_dir = []
amz_ad_report_full_dir = []


for i in range(len(amz_business_report_name)):
    amz_business_report_full_dir.append(os.path.join(amz_business_report_base_dir,amz_business_report_name[i]))

for i in range(len(amz_ad_report_name)):
    amz_ad_report_full_dir.append(os.path.join(amz_ad_report_base_dir,amz_business_report_name[i]))

print(amz_business_report_full_dir)
print(amz_ad_report_full_dir)
```

```
amz_business_report_name
```

```
business_report_DE = pd.read_csv(amz_business_report_full_dir[0])
business_report_DE['Country'] = amz_business_report_name[0][:2]

business_report_ES = pd.read_csv(amz_business_report_full_dir[1])
business_report_ES['Country'] = amz_business_report_name[1][:2]

business_report_FR = pd.read_csv(amz_business_report_full_dir[2])
business_report_FR['Country'] = amz_business_report_name[2][:2]

business_report_IT = pd.read_csv(amz_business_report_full_dir[3])
business_report_IT['Country'] = amz_business_report_name[3][:2]

business_report_NL = pd.read_csv(amz_business_report_full_dir[4])
business_report_NL['Country'] = amz_business_report_name[4][:2]

business_report_GB = pd.read_csv(amz_business_report_full_dir[5])
business_report_GB['Country'] = amz_business_report_name[5][:2]
```

```
business_report_FR
```

```
for i in range(len(business_report_GB['Ordered Product Sales'])):
    sale_string = business_report_GB['Ordered Product Sales'][i][1:]
    business_report_GB['Ordered Product Sales'][i] = float(sale_string)

business_report_GB['Ordered Product Sales']  
```

```
def currency (business_report_name, euro_currency):
    for i in range(len(business_report_name['Ordered Product Sales'])):
        currency_string = business_report_name['Ordered Product Sales'][i]
        currency_number = currency_string[1:]
        currency_float = float(currency_number)
        business_report_name['Ordered Product Sales'][i] = currency_float * euro_currency


currency(business_report_DE, 0.94)
currency(business_report_FR, 0.94)
currency(business_report_ES, 0.94)
currency(business_report_IT, 0.94)
currency(business_report_NL, 0.94)
```

```
columns_list = ['Date', 'Sessions', 'Ordered Product Sales','Total Order Items', 'Country']

business_report_ALL = pd.concat([business_report_DE[columns_list],
business_report_GB[columns_list],
business_report_FR[columns_list],
business_report_ES[columns_list],
business_report_IT[columns_list],
business_report_NL[columns_list]]
,axis = 0)

Session_country = pd.pivot_table(business_report_ALL, index = 'Date', columns='Country', values = 'Sessions', aggfunc = np.sum)
Session_country = Session_country[['UK', 'DE', 'FR', 'ES', 'IT', 'NL']]
Session_country
```

```
orders_country = pd.pivot_table(business_report_ALL, index = 'Date', columns='Country', values = 'Total Order Items', aggfunc = np.sum, margins= False)
orders_country = orders_country[['UK', 'DE', 'FR', 'ES', 'IT', 'NL']]
#orders_country =  orders_country.iloc[:len(orders_country.index)-1]
orders_country.fillna(0)
```

```
sales_country = pd.pivot_table(business_report_ALL, index = 'Date', columns='Country', values = 'Ordered Product Sales', aggfunc = np.sum)
sales_country = sales_country[['UK', 'DE', 'FR', 'ES', 'IT', 'NL']]
```

```
plt.rcParams['figure.figsize'] = [20, 10]

fig = plt.figure()
ax = Session_country.plot.bar(stacked= True)
#ax2 = ax.twinx()
#ax2.plot(orders_country['All'].values, marker = 'o', linestyle = '-', color='c')
plt.title('Amazon Session', fontsize = 25)
plt.rc('legend', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 18)

fig = plt.figure()
ax = sales_country.plot.bar(stacked= True)
plt.title('Amazon Sales(£)', fontsize = 25)
plt.rc('legend', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 18)
```
